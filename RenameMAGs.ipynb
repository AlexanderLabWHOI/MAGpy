# Rename contigs within bins based on the assembly group name

```python
from Bio import SeqIO
import glob 
import os
```

```python
os.mkdir('mags/renamed')
for file in glob.glob('mags/*fa'):
    fsplit = os.path.split(file)
    out = os.path.join(fsplit[0], 'renamed', fsplit[1])
    with open(file) as f, open(out, 'w') as o: 
        fname = file.split('/')[1].split('_')[0]
        records = SeqIO.parse(f, 'fasta')
        for record in records:
            newid = fname + '_' + record.id
            record.id = newid
            record.description = ''
            SeqIO.write(record, o, 'fasta')
```
